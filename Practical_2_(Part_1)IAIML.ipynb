```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
# 10 random integers between 0 and 50
x = np.random.randint(0 , 50 , 1000)
```

```python
print(x)
```

```python
# Positive Correlation with some noise
#y = x + np.random.normal(0 , 10 , 1000)
y = x + np.random.randint(0 , 10 , 1000)
```

```python
print(y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
plt.scatter(x , y)
plt.show()
# if both a and y are increasing aor decreasing then that is positive correlation else negative correlation. if it is a random then it is a zero correalation
```

```python
np.corrcoef(x , y)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'a' : np.random.randint(0 , 50 , 1000)})
df['b'] = df['a'] + np.random.randint(0 , 50 , 1000) # positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.randint(0 , 50 , 1000) # negatively correlated with 'a'
df['d'] = np.random.randint(0 , 50 , 1000) # not correlated with 'a'
```

```python
df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6 , 6))
plt.show()
```

```python
# Matrix plot
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show
```
